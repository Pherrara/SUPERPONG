```python
#turtle doc https://docs.python.org/3.3/library/turtle.html
#import turtle and create the main environment
import turtle

wn = turtle.Screen() #this defines the wn ("window") variable, capital S is important, it's a class
wn.title("SUPERPONG di Alessandro Michele Ferrara") #cool title
wn.bgcolor("black") #just background color
wn.bgpic("speedoflight.gif") #background image, MUST be a gif, cannot move tho
wn.setup(width = 1200, height = 900) #this makes it 4:3
wn.tracer(0) #this speeds up the game

#20 pixels = 1 unit 

#score and speed
startspeed = 0.2
initialspeed = 1
score1 = 0
score2 = 0

#left paddle:

paddle1 = turtle.Turtle() #capital T because it's a class
paddle1.speed(0) #this speeds up the paddle animation
paddle1.shape("circle") #this give it a square shape
paddle1.shapesize(stretch_wid = 12, stretch_len = 2) #the shape size, THE NAMES ARE JUST THESE, this means 40 pixels x 240 pixels
paddle1.color("#7F00FF") #just paddle color
paddle1.penup() #this way it doesn't draw a line when it moves
paddle1.goto(-500,0) #starting position

#right paddle:

paddle2 = turtle.Turtle() #capital T because it's a class
paddle2.speed(0) #this speeds up the paddle animation
paddle2.shape("circle") #this give it a square shape
paddle2.shapesize(stretch_wid = 12, stretch_len = 2) #the shape size, THE NAMES ARE JUST THESE
paddle2.color("#7F00FF") #just paddle color
paddle2.penup() #this way it doesn't draw a line when it moves
paddle2.goto(+500,0) #starting position DIFFERENT

#the ball:

ball = turtle.Turtle() #capital T because it's a class
ball.speed(0) #this speeds up the animation
ball.shape("circle") #this give it a square shape
ball.shapesize(stretch_wid = 2, stretch_len = 2) #the shape size
ball.color("yellow") #just the color
ball.penup() #this way it doesn't draw a line when it moves
ball.goto(0,0) #starting position DIFFERENT
ball.dx = startspeed #ball x speed
ball.dy = initialspeed #ball y speed

#score board
board = turtle.Turtle()
board.speed(0)
board.color("white")
board.penup()
board.hideturtle() #we don't wanna see the score board
board.goto(0, 250)
board.color("white")
board.write("Player A: 0    Player B: 0", align="center", font=("Courier", 24, "bold"))



#BOX
box1 = turtle.Turtle()
box1.speed(0)
box1.color("white")
box1.penup()
box1.goto(0, 450)
box1.shape("square") #this give it a square shape
box1.shapesize(stretch_wid = 0.1, stretch_len = 60) #the shape size, THE NAMES ARE JUST THESE

box2 = turtle.Turtle()
box2.speed(0)
box2.color("white")
box2.penup()
box2.goto(0, -450)
box2.shape("square") #this give it a square shape
box2.shapesize(stretch_wid = 0.1, stretch_len = 60) #the shape size, THE NAMES ARE JUST THESE

box3 = turtle.Turtle()
box3.speed(0)
box3.color("white")
box3.penup()
box3.goto(600, 0)
box3.shape("square") #this give it a square shape
box3.shapesize(stretch_wid = 45, stretch_len = 0.1) #the shape size, THE NAMES ARE JUST THESE

box4 = turtle.Turtle()
box4.speed(0)
box4.color("white")
box4.penup()
box4.goto(-600, 0)
box4.shape("square") #this give it a square shape
box4.shapesize(stretch_wid = 45, stretch_len = 0.1) #the shape size, THE NAMES ARE JUST THESE

#functions:

def paddle1_up():
    if paddle1.ycor() > 230: #this prevent further movement
        paddle1.sety(230)
    y = paddle1.ycor() #this sets the y var to the y coordinate of the object
    y += 100 #let's increment the y var
    paddle1.sety(y) #this sets the y coordinate to the y var
    
    
def paddle1_down():
    if paddle1.ycor() < -230: #this prevent further movement
        paddle1.sety(-230)
    y = paddle1.ycor() #this sets the y var to the y coordinate of the object
    y -= 100 #let's reduce the y var
    paddle1.sety(y) #this sets the y coordinate to the y var
    
def paddle1_right():
    if paddle1.xcor() > -200: #this prevents further movement
        paddle1.setx(-200)
    x = paddle1.xcor() #this sets the y var to the y coordinate of the object
    x += 40 #let's increment the y var
    paddle1.setx(x) #this sets the y coordinate to the y var
    
def paddle1_left():
    if paddle1.xcor() < -560: #this prevent further movement
        paddle1.setx(-560)
    x = paddle1.xcor() #this sets the y var to the y coordinate of the object
    x -= 40 #let's increment the y var
    paddle1.setx(x) #this sets the y coordinate to the y var    
    
def paddle2_up():
    if paddle2.ycor() > 230: #this prevent further movement
        paddle2.sety(230)
    y = paddle2.ycor() #this sets the y var to the y coordinate of the object
    y += 100 #let's increment the y var
    paddle2.sety(y) #this sets the y coordinate to the y var
    
def paddle2_down():
    if paddle2.ycor() < -230: #this prevent further movement
        paddle2.sety(-230)
    y = paddle2.ycor() #this sets the y var to the y coordinate of the object
    y -= 100 #let's reduce the y var
    paddle2.sety(y) #this sets the y coordinate to the y var
    
def paddle2_right():
    if paddle2.xcor() > 560: #this prevent further movement
        paddle2.setx(560)
    x = paddle2.xcor() #this sets the y var to the y coordinate of the object
    x += 40 #let's increment the y var
    paddle2.setx(x) #this sets the y coordinate to the y var
    
def paddle2_left():
    if paddle2.xcor() < 200: #this prevent further movement
        paddle2.setx(200)
    x = paddle2.xcor() #this sets the y var to the y coordinate of the object
    x -= 40 #let's increment the y var
    paddle2.setx(x) #this sets the y coordinate to the y var    
    
    
    
#KB bindings:

wn.listen() #receive inputs from kb in general
wn.onkeyrelease(paddle1_up, "w") #with the W key the left paddle goes up, just lowercase w works
wn.onkeyrelease(paddle1_down, "s") #with the S key the left paddle goes down, just lowercase s works
wn.onkeyrelease(paddle1_right, "d")
wn.onkeyrelease(paddle1_left, "a")
wn.onkeyrelease(paddle2_up, "Up") #with the Up arrow key the right paddle goes up
wn.onkeyrelease(paddle2_down, "Down") #with the Down arrow key the right paddle goes up
wn.onkeyrelease(paddle2_right, "Right")
wn.onkeyrelease(paddle2_left, "Left")



#main game loop
while True: #good old while True
    wn.update()
    
    #move the ball
    ball.setx(ball.xcor() + ball.dx) #this way the ball always moves along x
    ball.sety(ball.ycor() + ball.dy) #this way the ball always moves along y
    
    #bounce up and down
    if ball.ycor() > 430: 
        ball.sety(430)
        ball.dy *= -1
        
    if ball.ycor() < -430: 
        ball.sety(-430)
        ball.dy *= -1
        
    #return to center
    if ball.xcor() > 580: 
        paddle1.setx(-500)
        paddle2.setx(+500)
        ball.goto(0,0) #return to center
        ball.dx = initialspeed #initialize speed
        score1 += 1
        board.clear() #this cleans up the board
        board.color("white")
        board.write("Player A: {}    Player B: {}".format(score1,score2), align="center", font=("Courier", 24, "bold"))
        
    if ball.xcor() < -580: 
        paddle1.setx(-500)
        paddle2.setx(+500)
        ball.goto(0,0) #returns to center
        ball.dx = -initialspeed #initialize speed
        score2 += 1
        board.clear() #this cleans up the board
        board.color("white")
        board.write("Player A: {}    Player B: {}".format(score1,score2), align="center", font=("Courier", 24, "bold"))
        
    #paddles collisions
    if (ball.xcor() > paddle2.xcor() and ball.xcor() < paddle2.xcor() + 80) and (ball.ycor() < paddle2.ycor() + 120 and ball.ycor () > paddle2.ycor() - 120):
        ball.setx(paddle2.xcor())
        ball.dx *= -1
        paddle2.setx(paddle2.xcor()+10*abs(ball.dx))#backfire
        if paddle2.xcor() > 580: #this prevent further movement
            paddle2.setx(580)
        if abs(ball.dx) > 50: 
            while False: 
                nothing = 0
        else:
            ball.dx -= 0.5 #speed modifier
        
    if (ball.xcor() < paddle1.xcor() and ball.xcor() > paddle1.xcor() - 80) and (ball.ycor() < paddle1.ycor() + 120 and ball.ycor () > paddle1.ycor() - 120):
        ball.setx(paddle1.xcor())
        ball.dx *= -1
        paddle1.setx(paddle1.xcor()-10*abs(ball.dx)) #backfire
        if paddle1.xcor() < -580: #this prevent further movement
            paddle1.setx(-580)
        if abs(ball.dx) > 50: 
            while False: 
                nothing = 0
        else:
            ball.dx += 0.5 #speed modifier
        
```
